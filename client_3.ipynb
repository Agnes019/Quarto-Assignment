```python
import pandas as pd
from matplotlib import pyplot as plt

# Read CSV into pandas
data = pd.read_csv(r"unicef_indicator_2.csv")
data.head()
df = pd.DataFrame(data)

name = df['year'].head(12)
price = df['country'].head(12)

# Figure Size
fig = plt.figure(figsize =(10, 7))

# Horizontal Bar Plot
plt.bar(name[0:1960], price[0:1960])

# Show Plot
plt.show()
```
